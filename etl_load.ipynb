```python
# Importing libraries
import pandas as pd
from pathlib import Path

# Define file paths
output_dir = Path('loaded_data')
output_dir.mkdir(exist_ok=True)

# Output parquet paths
full_parquet = output_dir / 'full_data.parquet'
incremental_parquet = output_dir / 'incremental_data.parquet'
```

```python
# Loading the full transformed CSV
df_full = pd.read_csv("transformed_full.csv")

# Saving as Parquet
df_full.to_parquet(full_parquet, index=False)
```

```python
# Loading the incremental transformed CSV
df_incremental = pd.read_csv("transformed_incremental.csv")

# Saving as Parquet
df_incremental.to_parquet(incremental_parquet, index=False)
```

```python
# Reading back the saved Parquet files and showing previews
print("Full Data")
display(pd.read_parquet(full_parquet).head())

print("Incremental Data")
display(pd.read_parquet(incremental_parquet).head())

dddd
```
